#Data Types

#Numeric

```python
val1 = 10
print(val1)
print(type(val1))
print(sys.getsizeof(val1))
print(val1, "is integer?", isinstance(val1, int))
```

```python
val2 = 92.78
print(val2)
print(type(val2))
print(sys.getsizeof(val2))
print(val2, "is float?", isinstance(val2, float))
```

```python
val3 = 25 + 10j
print(val3)
print(type(val3))
print(sys.getsizeof(val3))
print(val3, "is complex?", isinstance(val3, complex))
```
