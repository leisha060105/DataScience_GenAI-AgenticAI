```python
x = 2

if x == 1:
    print("One")
elif x == 2:
    print("Two")
elif x == 3:
    print("Three")
elif x == 4:
    print("Four")

#this is more efficient, it will stop checking once it finds a match
```

```python
i = 1  #initialize the counter

while i <= 5:  #condition
    print('data science')
    i += 1  #increment the counter
```

```python
i = 1  #initialize the counter

while i <= 5:  #condition
    print('data science :', i)
    i += 1  #increment the counter
```

```python
i = 5  #initialize the counter

while i >= 1:  #condition
    print('data science :', i)
    i -= 1  #decrement the counter
```

```python
i = 1

while i <= 5:  #condition
    print('data science :', i)
    j = 1
    while j <= 4:
        print('technology')
        j += 1

    i += 1  #increment the counter
    print()
```
