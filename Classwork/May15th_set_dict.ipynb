Superset, subset, disjoint

```python
a2 = {1, 2, 3, 4, 5, 6, 7, 8, 9}
b2 = {3, 4, 5, 6, 7, 8}
c2 = {10, 20, 30, 40}
```

```python
a2.issuperset(b2)
```

```python
a2.isdisjoint(c2)
```

```python
a2.isdisjoint(b2)
```
